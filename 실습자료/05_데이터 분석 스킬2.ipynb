### 1.2. merge()를 활용한 수평 결합(병합) 
<br>
<img src="./data/20230827_161413.png"  width = 480 align="center" ></img>

- 특정한 KEY를 가지는 데이터세트간에 동일한 KEY를 기준으로 다른 변수를 붙이는 방법
- 같은 KEY를 가지는 다른 변수들이 옆으로 결합되기에 수평결합(병합)
- SQL의 JOIN, Excel의 VLOOKUP()같이 pandas의 **merge()** 를 활용하여 컬럼을 수평결합



```python
# 예제 데이터 불러오기
df_left  = pd.read_csv('data/data_left.csv')
df_right = pd.read_csv('data/data_right.csv')
```

```python
df_left
```

```python
df_right
```

<br>

#### how옵션을 통한 병합 방법 설정

> key를 활용한 병합에서는 일치하지 않는 key를 포함하는 기준에 따라 결과가 달라짐.

*merge()* 에서 `how=` 옵션을 활용해서 다음과 같은 데이터 결합 방법 지정 

+ `inner`: inner join. 양쪽다 일치하는 key만 포함
+ `left`:  left join. 왼쪽 데이터는 전부 포함하고 오른쪽 데이터는 일치하는 key만 포함
+ `right`: right join. 오른쪽 데이터는 전부 포함하고 왼쪽 데이터는 일치하는 key만 포함
+ `outer`: full outer join. 양쪽다 일치하지 않는 key까지 포함

```python
# merge()를 활용한 결합
pd.merge(df_left, df_right, how='inner', on='category')
```

```python
# left join
pd.merge(df_left, df_right, how='left', on='category')
```

```python
# right join
pd.merge(df_left, df_right, how='right', on='category')
```

```python
# full outer join
pd.merge(df_left, df_right, how='outer', on='category')
```

#### join()을 활용한 병합
> merge()와 같은 역할을 하지만 KEY는 반드시 인덱스

```python
df_left.set_index('category').join(df_right.set_index('category'), how = 'left')
```

## 2. 데이터 집계
pandas에는 다양한 방식으로 데이터를 합계, 평균과 같은 통계적으로 집계하는 함수를 제공

### 2.1. 수치형 집계값 계산

수치형 타입을 가지는 특정 컬럼에서 간단한 함수를 통해 통계적 집계 가능


```python
df_sp = pd.read_csv('./data/StudentsPerformance.csv')
df_sp.head()
```

```python
# mean() 함수를 활용하여 특정 컬럼의 평균 계산

df_sp['math score'].mean()
```

```python
# sum()으로 합계 계산

df_sp[['math score','reading score']].sum()
```

```python
# var()으로 분산 계산 
df_sp['math score'].var()
```

```python
# std()으로 표준편차계산
df_sp['math score'].std()
```

```python
# count()으로 개수 계산 (고유값 x)
df_sp['parental level of education'].count()
```



### 2.2. 그룹별 데이터 집계

> SQL 등에서 데이터 집계중 가장 활용도가 높은 연산은 *GROUPBY*.  
> pandas또한 *groupby()* 함수로 그룹화 하여 집계


```python
df_sp = pd.read_csv('./data/StudentsPerformance.csv')
df_sp.head()
```


#### *value_counts()* 함수로 특정 컬럼의 그룹별 개수 산출

```python
df_sp['parental level of education'].value_counts()
```

```python
# groupby()와 size()를 활용하여 동일 결과 산출
df_sp.groupby('parental level of education').size()
```

<br>

#### groupby() 함수를 활용한 그룹별 통계 수치
[pandas](https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html)
<img src="./data/20230827_163516.png"  width = 640 align="center" ></img>  
- groupby('그룹화컬럼')['집계컬럼'].통계함수()
- groupby('그룹화컬럼').통계함수()['집계컬럼']
- [['그룹화컬럼','집계컬럼']].groupby('그룹화컬럼').통계함수()

```python
# 그룹별 평균 계산
    #  by컬럼은 그룹화 하여 인덱스가됨
df_sp.groupby(by='race/ethnicity')['math score'].mean()
```

```python
# as_index=False 인자로 by컬럼을 인덱스가 아닌 컬럼으로 
df_sp.groupby('race/ethnicity', as_index=False)['math score'].mean()
```

```python
# 집계컬럼을 리스트로 묶어 넣으면 여러개의 컬럼으로 집계
df_sp.groupby(by='race/ethnicity')[['math score', 'reading score']].mean()
```

```python
df_sp.groupby(by='race/ethnicity')[['math score']].mean()
```

```python
# 동일한 결과의 다양한 표현
# 리스트등을 잘 활용하여 DataFrame유지
df_sp[['race/ethnicity','math score']].groupby('race/ethnicity', as_index=False).mean()
df_sp.groupby('race/ethnicity', as_index=False).mean()[['race/ethnicity','math score']]
```

```python
# 여러개 컬럼을 그룹화 하여 평균 계산
df_sp.groupby(['race/ethnicity', 'lunch'])['math score'].mean()
```

```python
# agg()를 활용한 그룹별 여러 집계 함수 적용
df_sp.groupby('race/ethnicity', as_index=False)['math score'].agg(['min','max','mean'])
```

```python
# agg()에 집계컬럼을 매핑하여 그룹화
    # 컬럼이 2개가 되는 단점 존재
df_sp.groupby('race/ethnicity', as_index=False).agg({'math score':['min','max'], 'reading score':['mean', 'median']})
```

### 2.2. 테이블 변환
- *pivot_table()*을 활용한 피벗테이블 생성
- *crosstab()*을 활용한 교차표를 생성

```python
df_cm = pd.read_csv('data/Churn_Modeling.csv')
df_cm
```

#### 2.2.1 *pivot_table()* 을 활용하여 그룹화와 집계를 동시에 테이블의 구조를 변형
[pandas](https://pandas.pydata.org/docs/getting_started/intro_tutorials/07_reshape_table_layout.html)
<img src="./data/20230827_164851.png"  width = 420 align="center" ></img>  

- 그룹화를 진행할 컬럼을 피벗테이블의 인덱스와 컬럼으로 설정
- 그룹별 집계를 산출할 단일 컬럼을 값(value)로 설정

```python
# pivot_table() 활용
#     index  : 그룹화 하여 인덱스가될 기존 컬럼명
#     columns: 그룹화 하여 컬럼이될 기존 컬럼명
#     values : 그룹별 집계가될 기존 컬럼명
#     aggfunc: 집계 함수
pd.pivot_table(df_cm, index='Geography', 
               columns='IsActiveMember', 
               values='CreditScore', aggfunc='mean')
```

```python
# numpy의 연산 함수 지원
import numpy as np

pd.pivot_table(df_cm, index='Geography', 
               columns='IsActiveMember', values='CreditScore', 
               aggfunc=np.mean)
```

```python
# 여러개의 인덱스 또는 컬럼으로 피벗
pd.pivot_table(df_cm, index=['Geography','HasCrCard'], 
               columns='IsActiveMember', 
               values='CreditScore', aggfunc='mean')
```

#### 2.2.2 *crosstab()* 을 활용하여 특정 두컬럼의 출현빈도의 백분율로 교차표 생성 가능

```python
# normalize='all' 전체에 대한 백분율
ctab = pd.crosstab(index= df_cm['Geography'], 
                   columns = df_cm['NumOfProducts'], 
                   normalize='all')
ctab
```

#### normalize(정규화) 인자를 활용하여 정규화(백분율) 축을 설정
> 기본값은 `'all'`로 전체데이터의 총합이 1인 정규화(백분율)로 나타남

```python
# normalize='index' 인덱스에 대한 정규화(백분율)
ctab = pd.crosstab(df_cm['Geography'], df_cm['NumOfProducts'], normalize='index')
ctab
```

```python
# normalize='columns' 컬럼에 대한 정규화
ctab = pd.crosstab(df_cm['Geography'], df_cm['NumOfProducts'], normalize='columns')
ctab
```

```python
# margins=True 인자로 전체 표현
ctab = pd.crosstab(df_cm['Geography'], df_cm['NumOfProducts'], normalize='all', margins=True)
ctab
```
