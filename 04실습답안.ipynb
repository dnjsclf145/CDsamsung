#### [실습]  

1. subway 데이터에서 승차총승객수와 하차총승객수를 더한 승하차총객수 컬럼을 만들기

2. 사용일자의 년월일을 연도로만 나타나게 바꾸기(문자열 그대로)

3. 등록일자의 년월일을 년.월.일 표기로 나타나게 바꾸기(datetime로 바꿔서)


    

```python
df_subway = pd.read_csv('./data/CARD_SUBWAY_MONTH_202107.csv', encoding='CP949')
df_subway['사용일자'] = df_subway['사용일자'].astype('str').apply(lambda x: f'{x[:4]} {x[4:6]} {x[6:]}')
df_subway['등록일자'] = df_subway['등록일자'].astype('str').apply(lambda x: f'{x[:4]} {x[4:6]} {x[6:]}')
df_subway.head()
```

```python
df_subway['승하차총승객수'] = df_subway['승차총승객수'] + df_subway['하차총승객수']
df_subway
```

```python
df_subway['사용일자'] = df_subway['사용일자'].apply(lambda x: x[:4])
df_subway
```

```python
df_subway['등록일자'] = df_subway['등록일자'].astype('datetime64')
df_subway['등록일자'] = df_subway['등록일자'].apply(lambda x: x.strftime('%Y.%m.%d'))
df_subway
```

#### [심화] 사용일자를 요일별로 바꾸고 주말, 평일 두개의 범주를 가지는 컬럼을 추가하기(고유값 출력하여 확인)
    datetime에서 요일은 일요일~토요일 을 0~6으로 표현

```python
def week(x):
    if int(x) in [1,2,3,4,5]:
        return '평일'
    else:
        return '주말'

df_subway['요일'] = df_subway['사용일자'].astype('datetime64').apply(lambda x: x.strftime('%w'))
df_subway['요일'] = df_subway['요일'].apply(week)
df_subway['요일'].unique()
```

#### [심화] 
1. df_nf에서 imdb_votes의 결측치를 0으로 채우고 높은순으로 정렬하세요
2. 1번 결과에서 apply를 이용하여 genres가 drama와 thriller 둘다 포함하고 있는 데이터만 인덱싱하세요

```python
import pandas as pd
df_nf = pd.read_csv('./data/titles.csv')
df_nf.head()
```

```python
df_nf = df_nf.fillna({'imdb_votes':0})
top_200 = df_nf.sort_values('imdb_votes', ascending = False).iloc[:200]
```

```python
def genres(x):
    if 'drama' in x and 'thriller' in x:
        return True
    else:
        return False
    
top_200[top_200['genres'].apply(genres)]
```

```python
def genres(x):
    x = x.replace('[','')
    x = x.replace(']','')
    x = x.replace("'",'')
    wl = x.split(',')
    wl = [w.strip() for w in wl]
    if 'drama' in wl and 'thriller' in wl:
        return True
    else:
        return False

df_nf = df_nf.fillna({'imdb_votes':0})
top_200 = df_nf.sort_values('imdb_votes', ascending = False).iloc[:200]
top_200[top_200['genres'].apply(genres)]
```
