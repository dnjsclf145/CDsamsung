```python
from pathlib import Path
file_dir = Path('./data/')
file_list  = file_dir.glob('CARD_*.csv')

df_list = []
for path in file_list:
    df_list.append(pd.read_csv(path, encoding='CP949'))
pd.concat(df_list, ignore_index = False)
```
