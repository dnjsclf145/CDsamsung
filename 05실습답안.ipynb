#### [심화] df_nf에서 국가별 'imdb_score', 'imdb_votes'의 평균을 나타내세요. 이때 국가는 production_countries에서 가장 먼저 나타나는 국가로 수정하여 그룹화 합니다.

```python
import pandas as pd
df_nf = pd.read_csv('./data/titles.csv')
df_nf
```

```python
def country(x):
    x = x.replace('[','')
    x = x.replace(']','')
    x = x.replace("'",'')
    wl = x.split(',')
    return wl[0]

df_nf['country'] = df_nf['production_countries'].apply(country)
nf_gp = df_nf.groupby('country')[['imdb_votes']].sum()
nf_gp['imdb_score'] = df_nf.groupby('country')['imdb_score'].mean()
nf_sort = nf_gp.sort_values('imdb_votes', ascending = False)
top_20 = nf_sort.iloc[:20]
top_20.sort_values('imdb_score', ascending = False)
```

#### [심화] math score를 상중하 범주로 바꾸어 race/ethnicity에 범주 백분율로 나타내세요.(상중하 범위는 자유)


```python
def score(x):
    if x < 50:
        return '하'
    elif x < 80:
        return '중'
    else:
        return '상'
    
df_sp['socre'] = df_sp['math score'].apply(score)
pd.crosstab(df_sp['race/ethnicity'], df_sp['socre'], normalize='columns')
```
